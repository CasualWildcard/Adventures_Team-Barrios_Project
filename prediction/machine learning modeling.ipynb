```python
import pandas as pd
from prophet import Prophet

import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv(r'C:\Users\cblick\Desktop\Project_Barrios\Adventures_Team-Barrios_Project\back-end\csvStorage\us_rs_weekly_consumable_gas_summary_20220102-20230903.csv')
data = data.copy(deep=True)
```

```python
dff=pd.DataFrame(pd.date_range(start='2022-01-02', end='2025-09-03',freq='W'),columns=['Date'])
dff['Date']=pd.to_datetime(dff['Date'])
```

```python
# dff['Adjusted O2 (kg)'] = 0
dff['Adjusted N2 (kg)'] = 0
```

```python
dff
```

```python
#data = data[['Date','USOS O2 (kg)','RS O2 (kg)','Adjusted O2 (kg)','Resupply O2 (kg)']].copy()
data = data[['Date','Adjusted N2 (kg)']].copy()
```

```python
date_column='Date'
pred_column='Adjusted N2 (kg)'
data[date_column]=pd.to_datetime(data[date_column])
#data['Resupply O2 (kg)'] = data['Resupply O2 (kg)'].fillna(0)
#data['Resupply N2 (kg)'] = data['Resupply N2 (kg)'].fillna(0)
#data['Resupply Air (kg)'] = data['Resupply Air (kg)'].fillna(0)
```

```python
data
```

```python
train_df = data.iloc[:-10]
test_df = data.iloc[-10:]
```

```python
train_df[pred_column].plot(figsize=(15,8), title=pred_column, fontsize=14, label='Train')
test_df[pred_column].plot(figsize=(15,8), title=pred_column, fontsize=14, label='Test')
plt.legend()
plt.grid()
plt.show()
```

```python
model = auto_timeseries(forecast_length=102,score_type='rmse', time_interval='W', model_type='ML') # model type=best
model.fit(traindata= data, ts_column=date_column, target=pred_column, cv=5)
```

```python
ddf = pd.concat([data,dff])
ddf
```

```python
model.get_leaderboard()
```

```python
future_predictions = model.predict(dff,model="ML")
```

```python
future_predictions = future_predictions.set_index(['date'])
data = data.set_index(['date'])
train_df = train_df.set_index([date_column])
test_df = test_df.set_index([date_column])
```
